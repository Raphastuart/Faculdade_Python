Este notebook tem como objetivo explicar de forma simples como funciona a tipagem de de variáveis na linguagem Python.

Python utiliza **Tipagem dinâmica**, o que significa que não precisamos declarar o tipo da varíavel antes de utilizá-las.

O próprio Python identifica automaticamente o tipo de dado.

## O que é Tipagem?

Tipagem é a forma como uma linguagem de programação classifica os tipos de dados de uma variável.

Os tipos mais utilizados em Python são:

- **int** -> números inteiros
- **float** -> números decimais
- **str** -> textos
- **bool** -> true ou false

Em python, o tipo da variável é definido automaticamente quando atribuímos um valor a ela.

A seguir veremos exemplos de cada tipo.

```python
a = 10
b = 3.14
c = "Python"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

## Tipagem dinâmica

Uma característica importante do Python é que **uma variável pode mudar de tipo durante o tempo de execução**.

Isso acontece porque Python usa **Tipagem dinâmica**.

```python
x = 10
print(type(x))

x = "Python"
print(type(x))
```

## Verificando o tipo de uma variável

Em python podemos verifica o tipo de uma variável utilizando a função **type()**

```python
nome = "Raphael"
idade = 23

print(type(nome))
print(type(idade))
```

## Conversão de tipos (Casting)

Às vezes precisamos converter um tipo de dado para outro.

Exemplo: transformar um número inteiro em decimal.

```python
numero = 10

print(type(numero))

numero = float(numero)

print(type(numero))
```

## Conclusão

Neste notebook aprendemos que:

- Python utiliza **Tipagem dinâmica**
- Não é necessário declarar o tipo das variáveis
- O tipo é difinido automaticamente
- Podemos verificar o tipo usando **type()**
- Também é possível converter tipos de dados usando **var = tipo(var)**
